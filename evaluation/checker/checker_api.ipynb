````python
import pandas as pd

competing_evaluation_prompt = """
I created a dataset of long situational questions, aiming to simulate the complex and personalized questions that human users may ask in daily life, and tested it with different models.
You will see the answers of 3 different models. Your task is to compare the answers of these 3 models and give the best model, the second best model, and the third best model in your opinion.
You need to score based on 4 aspects, 0-5 points, the higher the score, the better, and then give your reasons, and finally compare the answers of the three models.

Grading criteria
1) informativeness: the degree of information richness in the answer/the amount of useful information. If most of the answer is nonsense, the score will be lower. That is, the proportion of the part that the user will stay on after a glance
2) Completeness: whether the answer fully answers all aspects of the user's question. First split the query into several concepts, then check if the answer covers all the concepts.
3) Novelty: the degree of information contained in the answer that requires search and is not known by common sense. Think does the answer provide information that human users may not know or hard to think of without searching.
4) Actionability: whether the user can take specific actions based on the answer, the more specific the answer, the higher the score, and the answer with only abstract answers has a lower score

Follow the criteria rules:
1) score 0-5 for each of the four aspects for each model, and then give explanation of your score. 
2) To better generate convincing explanation, you should always cite a short but specific part of the answer or summarize the answer that supports your score.
3) When scoring, compare the answers of this model to the other models. If the answer of this model is not outstanding, the score should not be too high.

Finally, compare the answers of the three models and give the best model, the second best model, and the third best model in your opinion.
The following is an example.
{example}

Now begin:
"""

eval_user_prompt = """
Here are the three models' responses you will rate:
>>>Query>>>
{query}

>>>Answer from Model 1>>>
{model1_answer}

>>>Answer from Model 2>>>
{model2_answer}

>>>Answer from Model 3>>>
{model3_answer}

>>> Rating>>> Repeat the query, then give the answers of the three models respectively, and then give your rating
Follow the above format to rate the three models
Query: ...
Model 1: 
...
Model 2:
...
Model 3:
...

And at the end, output a csv format file with the following columns:
query, model, informativeness, completeness, novelty, actionability
Strictly follow the CSV format, you need to include the header, and note that in csv, query must be enclosed in quotes. 

### CSV Output
```
YOUR CSV OUTPUT HERE
```

Now begin:
>>> Rating>>>
"""
````

````python

import csv
import io
def csv_parser_append(llm_outout: str, output_file, model_name, eval_count):
    csv_output_only = llm_outout.split('```')[-2].strip()
    eval_res_dir = os.path.dirname(output_file)
    os.makedirs(eval_res_dir, exist_ok=True)

    if not os.path.exists(output_file):
        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['query', 'model_type', 'informativeness', 'completeness', 'novelty', 'actionability','model_name', 'eval_count'])

    with io.StringIO(csv_output_only) as f:
        reader = csv.reader(f)
        next(reader)
        with open(output_file, 'a', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            for row in reader:
                row.append(model_name)
                row.append(eval_count)
                writer.writerow(row)
````
